```python
# 1. outlier treatment
plt.boxplot(grouped_df['recency'])
```

```python
# two types of outliers:
# - statistical
# - domain specific
```

```python
# removing (statistical) outliers
# removing more than 95 and less than 5 percentile outliers.
Q1 = grouped_df.amount.quantile(0.05)   # basically it is 25th and 75th percentile 
Q3 = grouped_df.amount.quantile(0.95)
IQR = Q3 - Q1
print(IQR)
grouped_df = grouped_df[(grouped_df.amount >= Q1 - 1.5*IQR) & (grouped_df.amount <= Q3 + 1.5*IQR)]
grouped_df
# generally we only conside 25th-75th percentile box but 50 percent of data gets neglected
# so we are only removing 10% of data and other 5th perecntile - 95th perecentile we are keeping them.
```

```python
# outlier treatment for recency
Q1 = grouped_df.recency.quantile(0.05)
Q3 = grouped_df.recency.quantile(0.95)
IQR = Q3 - Q1
grouped_df = grouped_df[(grouped_df.recency >= Q1 - 1.5*IQR) & (grouped_df.recency <= Q3 + 1.5*IQR)]
grouped_df
```

```python
# outlier treatment for frequency
Q1 = grouped_df.frequency.quantile(0.05)
Q3 = grouped_df.frequency.quantile(0.95)
IQR = Q3 - Q1
grouped_df = grouped_df[(grouped_df.frequency >= Q1 - 1.5*IQR) & (grouped_df.frequency <= Q3 + 1.5*IQR)]
grouped_df
```

```python
# 2. rescaling
rfm_df = grouped_df[['amount', 'frequency', 'recency']]

# instantiate
scaler = StandardScaler()

# fit_transform
rfm_df_scaled = scaler.fit_transform(rfm_df)
rfm_df_scaled.shape
```

```python
from sklearn.neighbors import NearestNeighbors
from random import sample
from numpy.random import uniform
import numpy as np
from math import isnan
 
def hopkins(X):
    d = X.shape[1]
    #d = len(vars) # columns
    n = len(X) # rows
    m = int(0.1 * n) 
    nbrs = NearestNeighbors(n_neighbors=1).fit(X.values)
 
    rand_X = sample(range(0, n, 1), m)
 
    ujd = []
    wjd = []
    for j in range(0, m):
        u_dist, _ = nbrs.kneighbors(uniform(np.amin(X,axis=0),np.amax(X,axis=0),d).reshape(1, -1), 2, return_distance=True)
        ujd.append(u_dist[0][1])
        w_dist, _ = nbrs.kneighbors(X.iloc[rand_X[j]].values.reshape(1, -1), 2, return_distance=True)
        wjd.append(w_dist[0][1])
 
    H = sum(ujd) / (sum(ujd) + sum(wjd))
    if isnan(H):
        print(ujd, wjd)
        H = 0
 
    return H
```

```python
#First convert the numpy array that you have to a dataframe
rfm_df_scaled = pd.DataFrame(rfm_df_scaled)
rfm_df_scaled.columns = ['amount', 'frequency', 'recency']
rfm_df_scaled.head()
```

```python
#Use the Hopkins Statistic function by passing the above dataframe as a paramter
hopkins(rfm_df_scaled)
```

## Finding the Optimal Number of Clusters

### SSD

```python
# elbow-curve/SSD
ssd = []
range_n_clusters = [2, 3, 4, 5, 6, 7, 8]
for num_clusters in range_n_clusters:
    kmeans = KMeans(n_clusters=num_clusters, max_iter=50)
    kmeans.fit(rfm_df_scaled)
    
    ssd.append(kmeans.inertia_)
    
# plot the SSDs for each n_clusters
# ssd
plt.plot(ssd)
```

### Silhouette Analysis

$$\text{silhouette score}=\frac{p-q}{max(p,q)}$$

$p$ is the mean distance to the points in the nearest cluster that the data point is not a part of

$q$ is the mean intra-cluster distance to all the points in its own cluster.

* The value of the silhouette score range lies between -1 to 1. 

* A score closer to 1 indicates that the data point is very similar to other data points in the cluster, 

* A score closer to -1 indicates that the data point is not similar to the data points in its cluster.

```python
# silhouette analysis
range_n_clusters = [2, 3, 4, 5, 6, 7, 8]

for num_clusters in range_n_clusters:
    
    # intialise kmeans
    kmeans = KMeans(n_clusters=num_clusters, max_iter=50)
    kmeans.fit(rfm_df_scaled)
    
    cluster_labels = kmeans.labels_
    
    # silhouette score
    silhouette_avg = silhouette_score(rfm_df_scaled, cluster_labels)
    print("For n_clusters={0}, the silhouette score is {1}".format(num_clusters, silhouette_avg))
    
    
```

## Hierarchical Clustering

```python
rfm_df_scaled.head()
```

```python
grouped_df.head()
```

```python
# single linkage
mergings = linkage(rfm_df_scaled, method="single", metric='euclidean')
dendrogram(mergings)
plt.show()
```

```python
# complete linkage
mergings = linkage(rfm_df_scaled, method="complete", metric='euclidean')
dendrogram(mergings)
plt.show()
```

```python
# 3 clusters
cluster_labels = cut_tree(mergings, n_clusters=3).reshape(-1, )
cluster_labels
```

```python
# assign cluster labels
grouped_df['cluster_labels'] = cluster_labels
grouped_df.head()
```

```python
# plots
sns.boxplot(x='cluster_labels', y='recency', data=grouped_df)
```

```python
# plots
sns.boxplot(x='cluster_labels', y='frequency', data=grouped_df)
```

### When to use what? Kmeans or Hierarchical clustering

- Kmeans ,when data is of bigger size ,since kmeans is non-linear,its like plate and running circular not building up. so less space it takes.
- Hierarchical, when data is small size, it takes lot of ram coz it has to memorize all levels. if your data is on cloud and u have software u can use hierarchical on big data also.
- it is just the software u have. On ur laptop less space.k-means or hierarchical clustering algorithm depends on your hardware and the data that you are dealing with.

- **We can do hierarchical clustering and basis of business knowledge decide how many clusters we want , after that for these clusters we can find the centroid and take this centroid as initial cluster centroid for Kmeans algorithmn**
- so we are not only deciding no.of clusters but also centroids from hierarchical and giving it to k means algorithm

- Hierarchical clustering generally produces better clusters, but is more computationally intensive.
